```python
#Punto12
#Per ogni anno, calcolare l'importo totale dei prestiti. Ogni prestito che ha pianificato il tempo di scadenza 
#e il tempo di erogazione in diversi anni deve avere il suo importo distribuito proporzionalmente al numero di
#giorni in ogni anno. Ad esempio, un prestito con tempo di erogazione il 1 ° dicembre 2016, 
#scadenza prevista per il 30 gennaio 2018 e importo 5000 USD ha un importo di 5000 USD * 31 / (31 + 365 + 30) = 363,85 
#per il 2016, 5000USD * 365 / (31+ 365 + 30) = 4284.04 per il 2017 e 5000USD * 30 / (31 + 365 + 30) = 352.11 per il 2018.
```

```python
#In primo luogo, si impostano le due colonne 'disburse_time'e 'planned_expiration_time'come date,
#e si rimuove l'orario
T_prestiti['disburse_time'] = pd.to_datetime(T_prestiti['disburse_time'].dt.date) 
T_prestiti['planned_expiration_time'] = pd.to_datetime(T_prestiti['planned_expiration_time'].dt.date)
```

```python
#Si crea una colonna dei tempi di rimborso dove verranno considerati soltanto i giorni e si rimuove la vecchia colonna

T_prestiti['tempo_di_rimborso_senza_ore']= (T_prestiti['planned_expiration_time']) - (T_prestiti['disburse_time'])
del T_prestiti['tempo_di_rimborso']
```

```python
#Si crea un DF dove si ricerca la corrispondenza del valore year nelle colonne disburse_time e planned_expiration_time
#In questo modo, si andranno a ricercare tutti quei prestiti che scadono nell'anno in cui vengono concessi
scadenza_annuale = PMP[PMP['disburse_time'].dt.year == PMP['planned_expiration_time'].dt.year]
scadenza_annuale
```

```python
#Per quanto riguarda i prestiti che scadano in anni diversi da quelli in cui sono stati concessi,
#si crea un altro DF dove metterò tutti quei prestiti che presentanto (per quanto concerne l'anno) planned_expiration_time > disburse_time
scadenza_più_anni = scadenza_annuale = PMP[PMP['disburse_time'].dt.year < PMP['planned_expiration_time'].dt.year]
scadenza_più_anni
```

```python
#Si crea una nuova colonna che riporterà l'inzio dell'anno di ogni disburse_time.
#Si imposta ogni valore della data in disburse time come '01-01' al di fuori dell'anno
scadenza_più_anni["inizio_anno"] = scadenza_più_anni["disburse_time"].dt.year.apply(lambda x: dt.datetime(x, 1, 1))
scadenza_più_anni
```

```python
#Si applica lo stesso ragionamento per la fine dell'anno. Si riporta in ogni colonna 12-31 e si aggiunge 
#l'anno dalla colonna 'inzio_anno'
scadenza_più_anni["fine_anno"] = scadenza_più_anni["inizio_anno"].dt.year.apply(lambda x: dt.datetime(x, 12, 31))
scadenza_più_anni
```

```python
def giorni(row):
    if row['disburse_time'].year == row['inizio_anno'].year:
        return row['fine_anno'] - row['disburse_time']
    elif row['planned_expiration_time'].year == row['fine_anno'].year:
        return row['planned_expiration_time'] - row['inizio_anno']
    else:
        return 365
```

```python
#Si crea una riga per ogni prestito attraverso questa funzione.
#Ogni qual volta inizio_anno=disburse time si opera una sottrazione tra fine anno e disburse time, così da calcolare
#quanti giorni ci sono dalla concessione del prestito alla fine dell'anno. 
#Se questa condizione non dovesse essere verificata, allora si verifica se planned_expiration_time=fine_anno. 
#Se così fosse, si opera la differenza tra planned_expiration_time e inizio_anno.
#In questo modo si contano quanti giorni ci sono tra l'inzio dell'anno e la restituzione del prestito.
#Se anche questa condizione non fosse soddisfatta verrà considerato il valore 365 (l'anno intero).


```

```python
#Si aggiungono al df i risultati della funzione 'giorni'
scadenza_più_anni['giorni'] = scadenza_più_anni.apply(lambda x : giorni(x), axis=1)
```

```python
scadenza_più_anni
```

```python
#In questo passaggio si convertono i formati, così da poter effettuare le operazioni
scadenza_più_anni['giorni'] = pd.to_timedelta(scadenza_più_anni['giorni'])
scadenza_più_anni['giorni'] = scadenza_più_anni['giorni'].dt.days.astype('int16')
scadenza_più_anni['tempo_di_rimborso_senza_ore'] = scadenza_più_anni['tempo_di_rimborso_senza_ore'].dt.days.astype('int16')
scadenza_più_anni.head()
```

```python
#Si crea un'altra colonna dove si riporterà quanto ogni prestito vale per il singolo anno.
#Si andrà a dividere il totale del singolo prestito per la durata complessiva di questo e, poi, a moltiplicarlo per
#i giorni imputati ad ogni anno 
scadenza_più_anni['valore_per_annata'] = (scadenza_più_anni['loan_amount']/scadenza_più_anni['tempo_di_rimborso_senza_ore'])* scadenza_più_anni['giorni']
scadenza_più_anni
```

```python
#Attraverso una group_by, e poi una sum, si crea  un nuovo DF che riporterà il totale per il singolo anno
parziale1 = scadenza_più_anni.groupby(scadenza_più_anni['inizio_anno'].dt.year).sum()['valore_per_annata'].reset_index().rename(columns={'valore_per_annata' : 'valore_per_annata'})
parziale1
#il Df parziale1 si riferisce soltanto ai prestiti che scadano in annate differenti, ora si aggiungerà il valore dei prestiti 
#che scadano nello stesso anno in cui vengono concessi.
```

```python
parziale2=scadenza_annuale.groupby(scadenza_annuale['disburse_time'].dt.year).sum()['loan_amount'].reset_index().rename(columns={'disburse_time': 'inizio_anno','loan_amount' : 'valore_per_annata2'})
parziale2
#Si crea un nuovo df dove, attraverso una groupby e una sum, si somma il totale del valore dei prestiti con scadenza annuale.
#Si rinominano le colonne.
```

```python
#Si fa una merge e si crea un ultimo DF con i due precedenti parziali, utilizzando le chiave 'inizio_anno'
finale = pd.merge(parziale1, parziale2, on = 'inizio_anno')
finale
```
